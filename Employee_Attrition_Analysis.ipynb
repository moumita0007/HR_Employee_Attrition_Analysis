# Outliers detection

```python
numeric_data = df.select_dtypes(include = np.number)
for col in numeric_data.columns:
    Q1 = df[col].quantile(0.25)
    Q2 = df[col].quantile(0.50)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    LW = Q1 - 1.5 * IQR
    UW = Q3 + 1.5 * IQR
    outlier = ((df[col]<LW)| (df[col]>UW))
    if outlier.sum() > 0:
        print(f"{col} has outliers = {outlier.sum()}")
    else:
        print(f"{col} has no outliers")
```

# Outlier treatment

```python
# monthly income has 114 outliers
df['MonthlyIncome'] = df['MonthlyIncome'].clip(LW,UW)
```

```python
# number companies worked has 52 outliers
df['NumCompaniesWorked'] = df['NumCompaniesWorked'].clip(LW,UW)
```

```python
# StockOptionLevel has 85 outliers
df['StockOptionLevel'] = df['StockOptionLevel'].clip(LW,UW)
```

```python
# TotalWorkingYears has 63 outliers
df['TotalWorkingYears'] = df['TotalWorkingYears'].clip(LW,UW)
```

```python
# TrainingTimesLastYear has 238 outliers
df['TrainingTimesLastYear'] = df['TrainingTimesLastYear'].clip(LW,UW)
```

```python
# YearsAtCompany has 104 outliers
df['YearsAtCompany'] = df['YearsAtCompany'].clip(LW,UW)
```

```python
# YearsInCurrentRole has 21 outliers
df['YearsInCurrentRole'] = df['YearsInCurrentRole'].clip(LW,UW)
```

```python
# YearsSinceLastPromotion has 107 outliers
df['YearsSinceLastPromotion'] = df['YearsSinceLastPromotion'].clip(LW,UW)
```

```python
# Not giving any analytical insights
df.drop(columns= ['EmployeeCount','EmployeeNumber'],inplace=True)
```
